```python
# Data Visualization

import matplotlib.pyplot as plt
import seaborn as sns

# Now we set the style
sns.set(style='whitegrid')

# Scatter plot - SepalLength vs PetalLength colored by Species
sns.scatterplot(x='SepalLengthCm' , y='PetalLengthCm' , hue='Species', data=df)
plt.title('Sepal Length vs Petal Length')
plt.show()
```

```python
#Histogram for SepalLengthCm
sns.histplot(df['SepalLengthCm'], kde=True ,color='Blue' )

plt.title('Distribution of SepalLength')
plt.show()
```

```python
# Box plot for all numeric Features

plt.figure(figsize=(10,6))
sns.boxplot(data=df.iloc[:, 1:5])  # Skip 'Id' column
plt.title("Box Plot of Numeric Features")
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.show()
```
